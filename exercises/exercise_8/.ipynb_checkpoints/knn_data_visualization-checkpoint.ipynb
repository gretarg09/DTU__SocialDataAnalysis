# KNN data visualization


```python
# lets start by getting the data
import csv
import math
import matplotlib.pyplot as plt
from collections import defaultdict
from collections import Counter
from sklearn import tree

data = []

def process(IncidntNum,Category,Descript,DayOfWeek,Data,Time,PdDistrict,Resolution,Address,X,Y,Location,PdId):
    data.append({
            "IncidntNum":IncidntNum,
            "Category":Category,
            "Descript":Descript,
            "DayOfWeek":DayOfWeek,
            "Date":Date,
            "Time":Time,
            "PdDistrict":PdDistrict,
            "Resolution":Resolution,
            "Address":Address,
            "X":X,
            "Y":Y,
            "Location":Location,
            "PdId":PdId})

with open("SFPD_Incidents_-_from_1_January_2003.csv", "rb") as csvfile:
    reader = csv.DictReader(csvfile,delimiter=",")
    for row in reader:
        IncidntNum = row["IncidntNum"]
        Category = row["Category"]
        Descript = row["Descript"]
        DayOfWeek = row["DayOfWeek"]
        Date = row["Date"]
        Time = row["Time"]
        PdDistrict = row["PdDistrict"]
        Resolution = row["Resolution"]
        Address = row["Address"]
        X = row["X"]
        Y = row["Y"]
        Location = row["Location"]
        PdId = row["PdId"]
        process(IncidntNum,Category,Descript,DayOfWeek,Date,Time,PdDistrict,Resolution,Address,X,Y,Location,PdId)
```

```python
import pyglet
import geoplotlib
from geoplotlib.utils import BoundingBox

# Get the data on correct format
data_prostitution_location = {
    "lat":[], 
    "lon":[] 
}

for line in data:
    if line["Category"] == "PROSTITUTION":
        latitute,longitute = line["Location"].split(",")
        latitute = float(latitute[1:])
        longitute = float(longitute[:-1])
        if float(latitute) < 90.00:
            data_prostitution_location["lat"].append(latitute)
            data_prostitution_location["lon"].append(longitute)

# Plot the the prostituion data points up on geoplot
max_lat = max(data_prostitution_location["lat"]) 
min_lat = min(data_prostitution_location["lat"])
max_lon = max(data_prostitution_location["lon"])
min_lon = min(data_prostitution_location["lon"])

# Plotting up the image   
bbox = BoundingBox(north=max_lat, west=min_lon, south=min_lat, east=max_lon)
geoplotlib.set_bbox(bbox)
geoplotlib.kde(data_prostitution_location,1)
geoplotlib.inline() 
```

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

kmeans_error = []
kmeans_labels = []
iterate = [i+2 for i in range(5)]

kmeans_data =  {
    "locations": {"lat":[], "lon":[] },
    "labels_knn2" : [],
    "labels_knn3" : [],
    "labels_knn4" : [],
    "labels_knn5" : [],
    "labels_knn6" : []
}


# Get the data on correct format for sklearn.cluster.kmeans
location_input = np.array(zip(data_prostitution_location["lat"],data_prostitution_location["lon"]))

for i in iterate:
    kMeans_ = KMeans(n_clusters=i, random_state=0).fit(location_input)
    kmeans_data["labels_knn{}".format(i)] = list(kMeans_.labels_)
    kmeans_error.append(kMeans_.inertia_)
    kmeans_labels.append(kMeans_.labels_)

kmeans_data["locations"]=data_prostitution_location
kmeans_data3 =  {
    "knn2" : {
        "class0":{ "locations":[]},
        "class1":{ "locations":[]}      
    },
    "knn3" : {
        "class0":{ "locations":[]},
        "class1":{ "locations":[]},
        "class2":{ "locations":[]}
    },
    "knn4" : {
        "class0":{ "locations":[]},
        "class1":{ "locations":[]},
        "class2":{ "locations":[]},
        "class3":{ "locations":[]}, 
    },
    "knn5" : {
        "class0":{ "locations":[]},
        "class1":{ "locations":[]},
        "class2":{ "locations":[]}, 
        "class3":{ "locations":[]}, 
        "class4":{ "locations":[]} 
    },
    "knn6" : {
        "class0":{ "locations":[]},
        "class1":{ "locations":[]},
        "class2":{ "locations":[]},
        "class3":{ "locations":[]}, 
        "class4":{ "locations":[]}, 
        "class5":{ "locations":[]}
    },
}

for i,value in enumerate(kmeans_labels):
    number_of_classes =  i+2
    for x,class_nr in enumerate(kmeans_labels[i]):
        kmeans_data3["knn{}".format(number_of_classes)]["class{}".format(class_nr)]["locations"].append({"lat": location_input[x][0], "lon":location_input[x][1]})
```

```python
import json
with open('kmeans_data.json', 'w') as fp:
    json.dump(kmeans_data3, fp)
```
